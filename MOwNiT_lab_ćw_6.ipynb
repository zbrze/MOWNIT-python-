```python
def relative_error(val, real_val):
  return abs(val - real_val)/abs(real_val)
```

```python
def real_value(fx_integral, from_point, to_point):
  return abs(fx_integral(to_point) - fx_integral(from_point))
```

```python
import matplotlib.pyplot as plt
def draw(func, fx, from_point, to_point, n):
  points = np.linspace(from_point,to_point,n)
  if(func == 'r'):
    for i in range(1,len(points)):
            c = points[i-1]
            d = points[i]
            plt.plot([c,c],[0,fx((c+d)/2)],color='blue')
            plt.plot([d,d],[0,fx(d)],color='blue')
            plt.plot([c,d],[fx((c+d)/2),fx((c+d)/2)],color='blue')
  if(func == 't'):
    for i in range(1,len(points)):
            c = points[i-1]
            d = points[i]
            plt.plot([d,d],[0,fx(d)],color='green')
            plt.plot([c,c],[0,fx(c)],color='green')
            plt.plot([c,d],[fx(c),fx(d)],color='green')
  plt.plot(points, fx(points), 'r--')
  plt.show()
```

```python
import numpy as np
import time

def rectangular(fx, from_point, to_point, n):
  points = np.linspace(from_point,to_point,n)
  result = 0
  for i in range(1, n):
     result +=(points[i]-points[i-1])*fx((points[i]+points[i-1])/2)
  
 
  print("rectangular method ", n, "points, result:    ",result)
  return result

```

```python
def trapezoidal(fx, from_point, to_point, n):
  points = np.linspace(from_point,to_point,n)
  result = 0
  for i in range(1, n):
     result +=(points[i]-points[i-1])*(fx(points[i])+fx(points[i-1]))/2
  
  print("trapezoidial method ", n, "points, result:   ",result)
  return result
```

```python
def simpson(fx, from_point, to_point, n):
    if n % 2:
        raise ValueError("n must be even (received n=%d)" % n)
    points = np.linspace(from_point,to_point,n)
    h = (to_point - from_point) / n
    s = fx(from_point) + fx(to_point)
    points = np.linspace(from_point,to_point,n)
    for i in range(1, n, 2):
        s += 4 * fx(from_point + h*i)
    for i in range(2, n-1, 2):
        s += 2 * fx(from_point + h* i)
    print("simpson method ", n, "points, result:   ", s * h / 3)
    return s * h / 3
    
```

```python
myFuncs = [rectangular, trapezoidal, simpson]
n = [4, 20, 100]
ranges = [10, 1.5, 500]
for k in range (len(ranges)):
  print("range from 0 to ", ranges[k])
  print("real value: ", real_value(lambda x: 0.5*(x**2), 0, ranges[k]) )
  for j in range (len(n)):
    print("\n n = ", n[j])
    for i in range (0, 3):
      print("\n")
      start = time.time()
      print("relative error: ", relative_error(myFuncs[i](lambda x: x, 0, ranges[k], n[j]),real_value(lambda x: 0.5*(x**2), 0, ranges[k]) ))
      end = time.time()
      draw(str(myFuncs[i])[10], lambda x: x, 0, ranges[k], n[j])
      print("Function performance: ",end - start)
    print("\n\n\n")
  print("\n\n\n\n")
```
