```python
def real_value(fx_integral, from_point, to_point):
  return abs(fx_integral(to_point) - fx_integral(from_point))
```

```python
def relative_error(val, real_val):
  return abs(val - real_val)/abs(real_val)
```

```python
import numpy as np
from scipy.special.orthogonal import roots_legendre

def GaussLegendreQuadrature(f, a, b, degree):
  [x, w]= roots_legendre(degree+1)
  result = (b - a)* 0.5 * sum( w * f( (b - a)*0.5* x + (b + a) * 0.5 ) )
  print("Gauss - Legendre quadrature result:  ",result)
  return result
```

```python
from scipy.special.orthogonal import roots_hermite
def GaussHermiteQuadrature(f, a, b, degree):
  [x, w]= roots_hermite(degree+1)
  result = (b - a)* 0.5 * sum( w * f( (b - a)*0.5* x + (b + a) * 0.5 ) )
  print("Gauss - Hermite quadratuer result:  ",result)
  return result
```

```python
from scipy.special.orthogonal import roots_chebyt
def GaussChebytQuadrature(f, a, b, degree):
  [x, w]= roots_chebyt(degree+1)
  result = (b - a)* 0.5 * sum( w * f( (b - a)*0.5* x + (b + a) * 0.5 ) )
  print("Gauss - Chebytshev quadratuer result:  ",result)
  return result
```

```python
def rectangular(fx, from_point, to_point, n):
  points = np.linspace(from_point,to_point,n)
  result = 0
  for i in range(1, n):
     result +=(points[i]-points[i-1])*fx((points[i]+points[i-1])/2)
  print("Rectangular method result:  ",result)
  return result
```

```python
def trapezoidal(fx, from_point, to_point, n):
  points = np.linspace(from_point,to_point,n)
  result = 0
  for i in range(1, n):
     result +=(points[i]-points[i-1])*(fx(points[i])+fx(points[i-1]))/2
  print("Trapezoidal method result:  ",result)
  return result
```

```python
def simpson(fx, from_point, to_point, n):
    if n % 2:
       print("Simpson method result: degree is not even")
       return 0
    points = np.linspace(from_point,to_point,n)
    h = (to_point - from_point) / n
    s = fx(from_point) + fx(to_point)
    points = np.linspace(from_point,to_point,n)
    for i in range(1, n, 2):
        s += 4 * fx(from_point + h*i)
    for i in range(2, n-1, 2):
        s += 2 * fx(from_point + h* i)
    print("Simpson method result:  ", s * h / 3)
    return s * h / 3
```

```python
myFuncs = [rectangular, trapezoidal, simpson, GaussLegendreQuadrature, GaussHermiteQuadrature]
n = [2, 3, 4, 5]
a = 1
b = 10
func = lambda x: 3 * x**3 - 1
intFunc = lambda x: 0.75 * x**4 - x
print('\033[91m' +"calculated value:  ",real_value(intFunc, a, b))
print("\n")
for i in range (len(n)):
  print('\033[91m' + "degree:  ", n[i])
  for f in range (0,5):
      print('\033[94m'+"\n")
      print('\033[94m'+"relative error: ", relative_error(myFuncs[f](func, a, b, n[i]),real_value(intFunc, a, b) ))
      print("------------------------------------------------")
  print("\n\n")
```
