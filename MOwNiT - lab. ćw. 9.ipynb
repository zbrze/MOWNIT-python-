# Zadanie 1
### Zaimplemenuj metodę eliminacji Gaussa bez pivotingu i z pivotingiem dla układu równań o dowolnym rozmiarze.

```python
import numpy as np
from numpy.linalg import matrix_rank, det

def Gauss_basic(matrixA, matrixB):
  rows = matrixA.shape[0]
  for i in range (0, rows):
    for j in range (i+1, rows):
          quotient = float(matrixA[j][i])/matrixA[i][i]
          for k in range (i, rows):
            matrixA[j][k] -= float(matrixA[i][k]) * quotient
          matrixB[j] = float(matrixB[j]) - quotient * float(matrixB[i])
          matrixA[j][i] = 0
  solution = [0 for i in range(rows)]
  solution[rows - 1] = matrixB[rows- 1] / matrixA[rows - 1][rows - 1]
  for i in range(rows - 2, -1, -1):
        for j in range(i + 1, rows):
           matrixB[i] -= matrixA[i][j] * solution[j]
        solution[i] =  matrixB[i] / matrixA[i][i]
  return matrixA, solution
```

```python
def Gauss_pivoting(matrixA):
  n = len(matrixA)
  for i in range (n):

        # partial pivoting
        max_row = i #find row with the greatest number in column i
        counter = 1
        while i+counter<n:
            if abs(matrixA[i+counter,i]) > abs(matrixA[i,i]):
                max_row = i+counter
            counter +=1
        
        if i != max_row: #swap rows if max_row is different than i
            matrixA[max_row], matrixA[i] = matrixA[i], matrixA[max_row].copy()
           
        for j in range (i+1, n):
          quotient = float(matrixA[j][i])/matrixA[i][i]
          for k in range (i, n):
            matrixA[j][k] -= float(matrixA[i][k]) * quotient
          matrixA[j][i] = 0
  return matrixA
```

```python
import numpy as np
def j_pivot(row):
    for j in range(len(row)):
        if row[j] != 0:
            return j, True
    return 0, False

# returns the linear_combination row_i + t * row_j
def linear_combination(row_i, row_j, t):
    if len(row_i) != len(row_j):
        print('error: row_i and row_j must have the same length')
        exit(1)

    comb = [0] * len(row_i)
    for i in range(len(row_i)):
        comb[i] = row_i[i] + t * row_j[i]
    
    return comb

# returns the index of the first row, starting from top, with the leftmost pivot
def i_row_leftmost_pivot(matrix):
    i_leftmost_pivot, j_leftmost_pivot = 0, len(matrix[0])
    for i in range(len(matrix)):
        row = matrix[i]
        j_pivot_row, pivot_exists = j_pivot(row)
        if j_pivot_row < j_leftmost_pivot and pivot_exists:
            j_leftmost_pivot = j_pivot_row
            i_leftmost_pivot = i
    
    return i_leftmost_pivot



# main part of gaussian elimination algorithm
def elimination(matrix):
    
    i_to_swap = i_row_leftmost_pivot(matrix)
    matrix[0], matrix[i_to_swap] = matrix[i_to_swap], matrix[0]

    j_pivot_first_row, pivot_exists = j_pivot(matrix[0])

    if not pivot_exists:
        print('error: if first row has no pivot, and the first line row is the one with the leftmost pivot, matrix should be a stair matrix')
        exit(1)

    for i in range(1, len(matrix)):
        j_pivot_row_i, pivot_exists = j_pivot(matrix[i])

        if not pivot_exists:
            continue

        if j_pivot_row_i < j_pivot_first_row:
            print('error: first row should have the leftmost pivot')
            exit(1)
        
        if j_pivot_row_i == j_pivot_first_row:
            matrix[i] = linear_combination(matrix[i], matrix[0], -1 * matrix[i][j_pivot_row_i] / matrix[0][j_pivot_first_row])
    
    stair_matrix = [matrix[0]]
    stair_matrix.extend(elimination(matrix[1:]))
    return stair_matrix

# pretty print of the matrix
def print_matrix(matrix):
    for row in matrix:
        for elem in row:
            print('%.2f\t' % elem, end='')
        print()

def main():

    matrix = np.array([[453,  57,  27],
    [969, 504, 180],
    [102, 558, 685],
    [779, 583, 582],
    [799, 183, 581]]).reshape(5, 3) 
    matrix = elimination(matrix)
    print_matrix(matrix)

if __name__ == '__main__':
    main()
```

# Zadanie 3
### Podaj teorytyczną złożoność obliczeniową algorytmu eliminacji Gaussa. Przeprowadź testy wydajności swojego algorytmu sprawdzając jego działanie dla różnych rozmiarów macierzy (testy powinny być wykonane poza środowiskiem jupyter). Aby wygenerować układ równań, wygeneruj wektor rozwiązań i macierz współczynników losując wartości (skorzystaj z funkcji poznanych w Ćwiczeniu 2) i następnie oblicz wektor wyrazów wolnych.

złożoność obliczeniowa algorytmu eliminacji gaussa: O(n^3)

```python
import time
import random
import copy as cp
import matplotlib.pyplot as plt
import numpy as np

def Gauss_basic(matrixA, matrixB):
  rows = matrixA.shape[0]
  for i in range (0, rows):
    for j in range (i+1, rows):
          quotient = float(matrixA[j][i])/matrixA[i][i]
          for k in range (i, rows):
            matrixA[j][k] -= float(matrixA[i][k]) * quotient
          matrixB[j] = float(matrixB[j]) - quotient * float(matrixB[i])
          matrixA[j][i] = 0
  solution = [0 for i in range(rows)]
  solution[rows - 1] = matrixB[rows- 1] / matrixA[rows - 1][rows - 1]
  for i in range(rows - 2, -1, -1):
        for j in range(i + 1, rows):
           matrixB[i] -= matrixA[i][j] * solution[j]
        solution[i] =  matrixB[i] / matrixA[i][i]
  return matrixA, solution


def Gauss_pivoting(matrixA, matrixB):
    n = len(matrixA)
    for i in range(n):
        max_row = i 
        counter = 1
        while i + counter < n:
            if abs(matrixA[i + counter, i]) > abs(matrixA[i, i]):
                max_row = i + counter
            counter += 1

        if i != max_row: 
            matrixA[max_row], matrixA[i] = matrixA[i], matrixA[max_row].copy()
            matrixB[max_row], matrixB[i] = matrixB[i], matrixB[max_row].copy()
        for j in range(i + 1, n):
            quotient = float(matrixA[j][i]) / matrixA[i][i]
            for k in range(i, n):
                matrixA[j][k] -= float(matrixA[i][k]) * quotient
            matrixB[j] = float(matrixB[j]) - quotient * float(matrixB[i])
            matrixA[j][i] = 0
    solution = [0 for i in range(n)]
    solution[n - 1] = matrixB[n - 1] / matrixA[n - 1][n - 1]
    for i in range(n - 2, -1, -1):
        for j in range(i + 1, n):
            matrixB[i] -= matrixA[i][j] * solution[j]
        solution[i] = matrixB[i] / matrixA[i][i]
    return matrixA, solution


def performanceTests():
    n = random.randint(5, 100)
    matrixA = np.random.rand(n, n)
    matrixB = np.random.rand(n, 1)

    start1 = time.time()
    Gauss_basic(cp.deepcopy(matrixA), cp.deepcopy(matrixB))
    stop1 = time.time()
    basic_perf = stop1 - start1

    start2 = time.time()
    Gauss_pivoting(cp.deepcopy(matrixA), cp.deepcopy(matrixB))
    stop2 = time.time()
    pivot_perf = stop2 - start2

    return n, basic_perf, pivot_perf

if __name__ == '__main__':

    pivot = []
    basic = []

    for i in range (0,500):
        n_, b_, p_ = performanceTests()
        pivot.append((n_, p_))
        basic.append((n_, b_))

    pivot.sort(key=lambda tup: tup[0])
    basic.sort(key=lambda tup: tup[0])
    pivot_x = [x[0] for x in pivot]
    pivot_y = [x[1] for x in pivot]
    plt.title("Pomiary wydajności dla algorytmu z pivotingiem")
    plt.xlabel("Matrix size")
    plt.ylabel("Time")
    plt.scatter(pivot_x, pivot_y, 'bo')
    plt.savefig("wykres1")


    basic.sort(key=lambda tup: tup[0])
    basic.sort(key=lambda tup: tup[0])
    basic_x = [x[0] for x in basic]
    basic_y = [x[1] for x in basic]
    plt.title("Pomiary wydajności dla algorytmu bez pivotingu")
    plt.xlabel("Matrix size")
    plt.ylabel("Time")
    plt.scatter(basic_x, basic_y, 'or')
    plt.savefig("wykres2")

    plt.scatter(basic_x, basic_y, 'ro', label="without pivoting")
    plt.scatter(pivot_x, pivot_y, 'bo', label="pivoting")
    plt.title("Pomiary wydajności")
    plt.xlabel("Matrix size")
    plt.ylabel("Time")
    plt.plot(basic_x, basic_y, 'or')
    plt.legend()
    plt.savefig("wykres3")
```
